```python
df = pd.read_csv("marvel.csv", encoding='utf-8-sig')
df.columns = df.columns.str.strip().str.replace('\n', '').str.replace('\r', '')
for i, col in enumerate(df.columns):
    print(f"{i}: '{col}'")
df.drop(df.columns[6], axis=1, inplace=True)
print(df.head())







```

```python
df.head()
```

```python
df.rename(columns={'Movies_Appeared':'Movies'})
df.drop('Affiliation',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Power']=df['Power_Level'].apply(lambda x:x+101)
df
```

```python
df.drop(df.columns[3], axis=1, inplace=True)  
```

```python
df
```

```python
grouped_all = df.groupby('Power')['Movies_Appeared'].agg(['mean','sum','count'])
print(grouped_all)
```
