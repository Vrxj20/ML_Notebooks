```python
# Assignment 3: Array Operations

# Create two NumPy arrays of shape (3, 4) filled with random integers. Perform element-wise addition, subtraction, multiplication, and division.

A = np.random.randint(1, 11, (3, 4))
B = np.random.randint(1, 11, (3, 4))

print("Array A:\n", A)
print("Array B:\n", B)

# Element-wise operations
print("Addition:\n", A + B)
print("Subtraction:\n", A - B)
print("Multiplication:\n", A * B)
print("Division:\n", A / B)


#Create a NumPy array of shape (4, 4) with values from 1 to 16. Compute the row-wise and column-wise sum.

arr = np.arange(1, 17).reshape(4, 4)
print("Original array:\n", arr)

row_sum = np.sum(arr, axis=1)
col_sum = np.sum(arr, axis=0)
print("\nRow-wise sum:", row_sum)
print("\nColumn-wise sum:", col_sum)
```

```python
# Assignment 4: Statistical Operations
# Create a NumPy array of shape (5, 5) filled with random integers. Compute the mean, median, standard deviation,variance and normalised data of the array.

A = np.random.randint(1,101, size=(5,5))
print("\nOriginal Array: \n",A)

mean_val = np.mean(A)
median_val = np.median(A)
std_dev_val = np.std(A)
variance_val = np.var(A)

normalised_data = (A - mean_val) / std_dev_val

print("\n:Mean: ",mean_val)
print("\n:Median: ",median_val)
print("\n:Standard Deviation: ",std_dev_val)
print("\n:Variance: ",variance_val)
print("\n:Normalised Data: \n",normalised_data)



# Create a NumPy array of shape (3, 3) with values from 1 to 9. Normalize the array (i.e., scale the values to have a mean of 0 and a standard deviation of 1).
```

```python
# Assignment 5: Broadcasting
# Create a NumPy array of shape (3, 3) filled with random integers. Add a 1D array of shape (3,) to each row of the 2D array using broadcasting.

arr = np.random.randint(1,100, size=(3,3))
print("2D Array: \n",arr)

one_d = np.array([10,20,30])

result = one_d + arr
print("\nBroadcasting: \n",result)


# Create a NumPy array of shape (4, 4) filled with random integers. Subtract a 1D array of shape (4,) from each column of the 2D array using broadcasting.

arr1 = np.random.randint(1,100, size=(4,4))
print("\n 2D Array: \n",arr1)

one_dim = np.array([1,2,3,4])
result1 = arr1 - one_dim
print("\n Broadcasting: \n",result1)
```

```python
# Assignment 6: Linear Algebra

# Create a NumPy array of shape (3, 3) representing a matrix. Compute its determinant, inverse, and eigenvalues.
arr = np.array([
    [1,2,3],
    [2,3,9],
    [5,8,7]
])

deter = np.linalg.det(arr)
inverse = np.linalg.inv(arr)
eigenvalues, eigenvectors = np.linalg.eig(arr)


print("\nDETERMINANT:",deter)
print("\nINVERSE: \n",inverse)
print("\nEIGENVALUES:\n", eigenvalues)
print("\nEIGENVECTORS:\n", eigenvectors)



# Create two NumPy arrays of shape (2, 3) and (3, 2). Perform matrix multiplication on these arrays.

A = np.array([
    [1,2,3],
    [4,5,6]
])

B = np.array([
    [1,2],
    [3,4],
    [5,6]
])

matrix_multiplication = A @ B
print("\nMatrix multiplication: \n",matrix_multiplication)
```

```python
# Assignment 9: Structured Arrays
# Create a structured array with fields 'name' (string), 'age' (integer), and 'weight' (float). Add some data and sort the array by age.

dt = np.dtype([                                         #created a custom data type 
    ('name', 'U10'),
    ('age','i4'),
    ('weight','f4')
])

data = np.array([                                               
    ("Tony",25,68.5),
    ("Steve",320,78.9),
    ("Peter",22,63.00)
],dtype=dt)

sorted_data  = np.sort(data,order='age')
print("Sorted data by age: \n",sorted_data)



# Create a structured array with fields 'x' and 'y' (both integers). Add some data and compute the Euclidean distance between each pair of points.

dt1 = np.dtype([
    ('x','i4'),
    ('y','i4')
])

data = np.array([
    (1,2),
    (-8,20),
    (0,9),
    (16,-5)
],dtype=dt1)

points = np.column_stack((data['x'],data['y']))
distance = np.array([[np.linalg.norm(p1-p2) for p2 in points] for p1 in points])

np.set_printoptions(precision=2, suppress=True)
print("\nEuclidean distance between each pair is: \n",distance)
```
