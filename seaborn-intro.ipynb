```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset("tips")
data
```

```python
# SCATTER PLOT 

sns.scatterplot(x="tip",y="total_bill",data=data)
plt.title("Scatter Plot of Revenue w.r.t Days of a week")
plt.show()

```

```python
# lINE PLOT

sns.lineplot(x="size",y="total_bill",
             data=data,
             color="mediumvioletred")
plt.title("Total Bill with size")
plt.show()
```

```python
# CATEGORICAL PLOTS

#BAR PLOT
sns.barplot(x="total_bill",y="sex",data=data,
            color="crimson")
plt.title("Total Bill according to sex")
plt.show()
```

```python
#BAR PLOT

sns.barplot(x="day",y="total_bill",data=data)
```

```python
#VIOLIN PLOT
sns.violinplot(x="day",y="total_bill",data=data)
```

```python
#HISTOGRAM
sns.histplot(data['total_bill'],bins=15,kde=True) #bins = no.of bars || kde = Curve
```

```python
#KDE PLOT
sns.kdeplot(data['total_bill'],fill=True,color="orange")
```

```python
#PAIR PLOTS
sns.pairplot(data)
```

```python
#HEATMAPS
corr=data[['total_bill','tip','size']].corr()
corr
sns.heatmap(corr,annot=True,cmap='rocket')
```
